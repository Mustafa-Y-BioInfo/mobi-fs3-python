## Intensity clipping

First we define some fixed thresholds $T_1$, $T_2$:

```python
T1 = 0.31
T2 = 0.59
```

### Using `ndarray.clip` method

**<a id="#">Task 2.3.1.</a>** Enter your solution below:

```python
img_clipped =img.clip(T1,T2)

plt.figure(figsize=(25, 5))

plt.subplot(1, 3, 1)
plt.title('orginal')
plt.imshow(img, 'gray')
plt.colorbar()

plt.subplot(1, 3, 2)
plt.title('clipped')
plt.imshow(img_clipped, 'gray')
plt.colorbar()
```

### Not using the `ndarray.clip` method

**<a id="#">Task 2.3.2 (a).</a>** Enter your solution below:

```python
manuell_clip_1 = img.copy()
i=0 # row
r=0 # col
for i in range (0,768):
    for r in range (0,1024):
        if manuell_clip_1[i,r]<T1:
            manuell_clip_1[i,r]=T1
        elif manuell_clip_1[i,r]>T2:
            manuell_clip_1[i,r]=T2
        
plt.figure(figsize=(25,5))

plt.subplot(1, 3, 1)
plt.title('manuell_clip_1')
plt.imshow(manuell_clip_1, 'gray')
plt.colorbar()

plt.subplot(1, 3, 2)
plt.title('clipped')
plt.imshow(img_clipped, 'gray')
plt.colorbar()
```

**<a id="#">Task 2.3.2 (b).</a>** Enter your solution below:

```python
manuell_clip_2 = img.copy()
for row, col in numpy.ndindex(img.shape): 
    if manuell_clip_2[row, col]<T1:
        manuell_clip_2[row, col]=T1
    elif manuell_clip_2[row, col]>T2:
        manuell_clip_2[row, col]=T2

plt.figure(figsize=(25,5))

plt.subplot(1,3,1)
plt.title('manuell_clipped_2')
plt.imshow(manuell_clip_2, 'gray')
plt.colorbar()

plt.subplot(1, 3, 2)
plt.title('clipped')
plt.imshow(img_clipped, 'gray')
plt.colorbar()
```
